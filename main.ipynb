### Вывод
 #### - для cat1 - cat72 используем OrdinalEncoder
 #### - для остальных категориальных oneHotEncoder
 #### - для cont исопьзуем standardScaler

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
numeric_pipeline = make_pipeline(
    StandardScaler()
)
```

```python
binary_pipeline = make_pipeline(
    OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1, dtype=np.int8)
) 
```

```python
multi_cat_pipeline = make_pipeline(
    OneHotEncoder(
        handle_unknown='ignore',
        sparse_output=False,
        dtype=np.int8
    )
)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, num_features),
        ('bin_cat', binary_pipeline, binary_cat_features), 
        ('multi_cat', multi_cat_pipeline, multi_cat_features)
    ],
    remainder='passthrough'
)
```

```python
preprocessor
```

```python
X_train_df = df_train.drop(['loss'], axis=1)
```

```python
y_train_df = df_train['loss']
```

```python
X_processed = preprocessor.fit_transform(X_train_df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y_train_df, 
    test_size=0.1, 
    random_state=42
)
```

```python
print(f'train - {X_train.shape}')
print(f'test - {X_test.shape}')
```

### Задние 5 
**Строите полносвязную модель keras**

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, InputLayer
from keras.callbacks import EarlyStopping
from tensorflow.keras.layers import BatchNormalization

from keras.metrics import RootMeanSquaredError
```

```python
tf.random.set_seed(42)
np.random.seed(42)
```

```python
input_shape = X_train.shape[1]
```

```python
model = Sequential()

model.add(InputLayer(shape=(input_shape,))) 

model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(32, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mae', metrics=['mae'])
```

```python
early_stop = EarlyStopping(
    monitor='val_loss', 
    patience=20, 
    restore_best_weights=True,
    verbose=1
)
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=100,
    validation_split=0.2,
    verbose=0,
    callbacks=[early_stop]
)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Обучение')
plt.plot(history.history['val_loss'], label='Валидация')
plt.title('График Loss на обучении и валидации')
plt.xlabel('Эпоха')
plt.ylabel('Loss (MAE)')
plt.legend()
plt.grid(True)
plt.show()
```

### Проверка на тестовой выборке

```python
y_test_pred = model.predict(X_test).flatten()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_test, y_test_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
```

```python
print(f"MAE на тестовых данных: {mae:.4f}")
print(f"RMSE на тестовых данных: {rmse:.4f}")
```
